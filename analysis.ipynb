```python
import pandas as pd

data = pd.read_csv("fatigue-data.csv")
len(data)
data.head()
data.columns

#data.columns = ["timestamp","short_content_time","screen_time", "content_type","before_sleep_usage",
#"task_switching","breaks","sleep_duration","focus_rating","
len(data.columns)
for i, col in enumerate(data.columns):
    print(i, col)


data.rename(columns={
"Timestamp": "timestamp",
  "1. About how much time do you spend daily on short-form content (like reels, shorts, quick scrolling)?": "short_content_time",
 "Enter your average daily screen time (in hours) To check your screen time:\nAndroid: Go to Settings → Digital Wellbeing & Parental Controls → Dashboard\niPhone: Go to Settings → Screen Time → See All Activity": "screen_time",
   "What is your primary form of content consumed?":"content_type",
   "Do you engage with any digital content within 30 minutes before sleep?":"before_sleep",
    "How often do you switch between apps/tasks while using your device?":"task_switching",
    "How often do you take intentional breaks during screen usage?":"breaks",
    "What is your average sleep duration?":"sleep_duration",
    "Rate your ability to maintain focus on a single task (1–10)":"focus_rating",
    "How often do you feel mentally fatigued during the day?":"fatigue_level",
    "How easily do you get distracted while working/studying?":"distraction_level",
    "After periods of high screen usage, how does your concentration change?":"concentration_change"
},inplace=True)
data.columns

data.rename(columns={data.columns[2]: "screen_time"},inplace=True)
data.columns

#analysing the rwlationships between variables
pd.crosstab(data["content_type"],data["fatigue_level"])
#pd.crosstab(data["screen_time"],data["fat"])
pd.crosstab(data["screen_time"],data["focus_rating"])

#cleaning up screen time column
import re
def get_new_hours(value):
    value=str(value)
   ## extract_num=re.findall('\d+', value)
    new_num=re.findall(r'\d+', value)

    if new_num:
        return int(new_num[0])
    else:
        return None
    
data["screen_time"]=data["screen_time"].apply(get_new_hours)
data["screen_time"].head()

pd.crosstab(data["screen_time"],data["focus_rating"])
data.groupby("screen_time")["focus_rating"].mean()

col_names=["focus_rating","sleep_duration","task_switching","breaks","fatigue_level","distraction_level"]
##for columns in col_names:
def conv_to_num(x):
     try:
        return int(x)
     except:
        return None
for col in col_names:
    data[col] = data[col].apply(conv_to_num)
data.head()
print(len(data))
print(data.shape)
missing_values=data.isnull().sum()
print(missing_values)

cols = data.columns

data.rename(columns={
    cols[0]: "timestamp",
    cols[1]: "short_content_time",
    cols[2]: "screen_time",
    cols[3]: "content_type",
    cols[4]: "before_sleep",
    cols[5]: "task_switching",
    cols[6]: "breaks",
    cols[7]: "sleep_duration",
    cols[8]: "focus_rating",
    cols[9]: "fatigue_level",
    cols[10]: "distraction_level",
    cols[11]: "concentration_change"
}, inplace=True)
```
